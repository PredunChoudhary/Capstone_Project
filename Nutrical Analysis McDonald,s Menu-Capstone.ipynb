***6.Data Visualization:***

```python
## Identify trends and patterns
# Creating histograms plot for Visualisation

plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=40, kde=True)
plt.title('Distribution of Calorie Counts')
plt.show()
```

This code creates a histogram plot of calorie counts from a dataset. It helps visualize the distribution of calorie counts and identify trends or patterns in the data.
1. X axis-the range of calorie counts in the dataset
2. Y axis- the calorie counts that fall within each bin.



```python
import pandas as pd
df = pd.read_csv("Nutrical Dataset.csv")
df.head()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
average_nutrition = df.groupby(['Category'], as_index=False)[['Sugars','Protein', 'Calories', 'Carbohydrates']].mean()
```

```python
# Creating a barchart for calorie distribution 

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Calories',hue='Category', data=average_nutrition, palette='rainbow')
plt.title('Calories Content Across Menu')
plt.xlabel('Menu Categories')
plt.ylabel('Calories (g)')
plt.xticks(rotation=45,ha='right')
plt.legend(title='Calories', loc='upper right')
plt.show()
```

1.We have created a bar chart to visualize the calorie content of various menu categories. 

2.From the above bar plot of items, we can see that ***chicken and Fish*** has highest amount of Calories and ***Beverages*** has lowest.

3.Salads has more Calories than Desserts.



```python
# Box plot for distribution of calorie counts
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Calories',hue='Category', data=df, palette='rainbow')
plt.title('Box Plot of Calorie Counts Across Menu Categories')
plt.xlabel('Menu Categories')
plt.ylabel('Calorie Counts')
plt.xticks(rotation=45, ha='right')
plt.show()
```

1.We have created a box plot to visualize the distribution of calorie counts for various menu categories. This box plot provides an overview of the spread of calorie counts for each menu category and also helps to identify any potential outliers.

2.***hue*** is used to distinguish between different menu categories, allowing users to easily compare the calorie distribution across different menu types

3.Above box plot shows that ***Chicken and fish*** has highest amount of Calories and salads has more Calories than desserts.

4.All the plot( Histogram,Bar plot & Box plot) indicates the same thing about the nutrients distribution in the menu categories.


```python
# Creating a bar chart for nutritional content Protein

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Protein',hue='Protein', data=average_nutrition, palette='magma')
plt.title('Protein Content Across Menu')
plt.xlabel('Menu Categories')
plt.ylabel('Protein (g)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Protein', loc='upper right')  
plt.show()
```

1. We have created a bar chart to visualize the calorie content of various menu categories

2. From the bar chart we can clearly see that in menu categories ***Chicken and fish**** has highest protein content.

3. ***Beverages*** has lowest protein content.

4. ***Beef & Pork*** has less protein content than ***Chicken& Fish***.

```python
# Create a box plot for nutritional content Carbohydrates

plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Carbohydrates',hue='Category', data=df, palette='plasma')
plt.title('Box Plot of Carbohydrates Counts Across Menu Categories')
plt.xlabel('Menu Categories')
plt.ylabel('Carbohydrates Counts')
plt.xticks(rotation=45, ha='left' )
plt.show()
```
